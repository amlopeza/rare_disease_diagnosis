### Summary

**Dataset:** `Evaluation.csv` — 6,915 clinical case summaries spanning **1,012 unique rare diseases**.

**Exploration performed:**
- Loaded and inspected shape, columns, dtypes, and basic statistics
- Checked missing values across all columns — `Synonyms` (2,070), `OMIM` (2,378), `GARD` (1,503) had the most nulls; `CaseSummary` and `Disease` had zero
- Computed `text_length` for each case summary (range: 6–44,156 chars, median ~1,776)
- Identified the shortest summaries (6 chars) as placeholder text ("Case 1")
- Examined disease synonym coverage

**Key findings:**
1. The dataset is highly multi-class: 1,012 disease labels across 6,915 samples (~6.8 samples per disease on average)
2. Case summaries vary widely in length (6 to 44K characters), indicating heterogeneous data quality
3. ~30% of rows lack synonyms; medical coding columns (ICD-10, ICD-11, OMIM, GARD) also have notable missingness
4. `CaseSummary` and `Disease` are fully populated — no nulls

**Cleaning steps:**
- Selected a subset of columns for downstream use
- Filled missing `Synonyms` with `"Not available"`
- Label-encoded the `Disease` column into `category_encoded`

**Columns in `cleaned_data.csv`:**

| Column | Description |
|---|---|
| `CaseSummary` | Clinical case narrative (text) |
| `Disease` | Rare disease name (target label) |
| `text_length` | Character length of `CaseSummary` |
| `Synonyms` | Disease synonyms (nulls filled with "Not available") |
| `category_encoded` | Integer-encoded disease label (0–1011) |
